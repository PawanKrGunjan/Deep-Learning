```python
import math
import numpy as np

import matplotlib.pyplot as plt
```

### Step function

```python
def step(x):
    if x>=0:
        return 1
    else:
        return 0
```

```python
step(0.5)
```

```python
x = np.linspace(-10,10,100000)
y = np.array([step(i) for i in x])

plt.figure(dpi=100)
plt.axvline(x = 0, color = 'g', linestyle = '-')
plt.axhline(y = 0, color = 'g', linestyle = '-')
plt.plot(x,y, color = 'r')
plt.grid(True)
plt.title('Step Function')
plt.show()
```

### Sigmoid Function

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(35)
```

```python
x = np.linspace(-10,10,100000)
```

```python
x = np.linspace(-10,10,100000)
y = np.array([sigmoid(i) for i in x])

plt.figure(dpi=100)
plt.axvline(x = 0, color = 'g', linestyle = '-')
plt.axhline(y = 0, color = 'g', linestyle = '-')
plt.plot(x,y, color = 'r')
plt.grid(True)
plt.title('Sigmoid Function')
plt.show()
```

### Tanh Function

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-10)
```

```python
x = np.linspace(-10,10,100000)
y = np.array([tanh(i) for i in x])

plt.figure(dpi=100)
plt.axvline(x = 0, color = 'g', linestyle = '-')
plt.axhline(y = 0, color = 'g', linestyle = '-')
plt.plot(x,y, color = 'r')

plt.grid(True)
plt.title('Tanh Function')
plt.show()
```

### Relu function

```python
def relu(x):
    return max(0,x)
```

```python
relu(-10)
```

```python
x = np.linspace(-10,10,100000)
y = np.array([relu(i) for i in x])

plt.figure(dpi=100)
plt.axvline(x = 0, color = 'g', linestyle = '-')
plt.axhline(y = 0, color = 'g', linestyle = '-')
plt.plot(x,y, color = 'r')

plt.grid(True)
plt.title('Relu Function')
plt.show()
```

### Leaky Relu Function

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(-500)
```

```python
x = np.linspace(-10,10,100000)
y = np.array([leaky_relu(i) for i in x])

plt.figure(dpi=100)
plt.axvline(x = 0, color = 'g', linestyle = '-')
plt.axhline(y = 0, color = 'g', linestyle = '-')
plt.plot(x,y, color = 'r')

plt.grid(True)
plt.title('Leaky Relu Function')
plt.show()
```

### Exponent Linear unit function

```python
def elu(x):
    return max(0.1*(math.exp(x)-1),x)
```

```python
elu(0.005)
```

```python
x = np.linspace(-10,10,100000)
y = np.array([elu(i) for i in x])

plt.figure(dpi=100)
plt.axvline(x = 0, color = 'g', linestyle = '-')
plt.axhline(y = 0, color = 'g', linestyle = '-')
plt.plot(x,y, color = 'r')

plt.grid(True)
plt.title("Exponent Linear unit")
plt.show()
```

### Rectifier 'or' Softplus Function

```python
def softplus(x):
    return math.log(1+math.exp(x))
```

```python
softplus(10)
```

```python
x = np.linspace(-10,10,100000)
y = np.array([softplus(i) for i in x])

plt.figure(dpi=100)
plt.axvline(x = 0, color = 'g', linestyle = '-')
plt.axhline(y = 0, color = 'g', linestyle = '-')
plt.plot(x,y, color = 'r')

plt.grid(True)
plt.title("Rectifier 'or' Softplus Function")
plt.show()
```

### Softmax function

```python
def softmax(x):
    return np.exp(x) / float(sum(np.exp(x)))
```

```python
x = np.linspace(-10,-10,5)
softmax(x)
```

```python
x = np.linspace(-10,10,30)
y = softmax(x)
print('x :',x)
print('y :',y)

plt.figure(dpi=100)
plt.axvline(x = 0, color = 'g', linestyle = '-')
plt.axhline(y = 0, color = 'g', linestyle = '-')
plt.plot(x,y, color = 'r')

plt.grid(True)
plt.title("Softmax Function")
plt.show()
```
